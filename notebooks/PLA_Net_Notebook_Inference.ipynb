## ⚛️ PLA-Net

Run inference using the gradio client for PLA-Net.


#### Inference

`protein_id`: protein id from the AD dataset


`ligand_smile`: SMILES in string, to input multiple separate them with `:`





```python
!pip install gradio_client
```

```python
from gradio_client import Client

client = Client("juliocesar-io/PLA-Net")
result = client.predict(
		protein_id="ada",
		ligand_smile="Cn4c(CCC(=O)Nc3ccc2ccn(CC[C@H](CO)n1cnc(C(N)=O)c1)c2c3)nc5ccccc45:OCCCCCn1cnc2C(O)CN=CNc12:Nc4nc(c1ccco1)c3ncn(C(=O)NCCc2ccccc2)c3n4",
		api_name="/run_inference"
)
```

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame(result['data'], columns=result['headers'])

# Display DataFrame
df
```
